```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregando o arquivo CSV
netflix_df = pd.read_csv('/content/sample_data/netflix_data.csv',index_col = 0)

netflix_df
```

```python
# Filtrando os TV Show
netflix_subset = netflix_df[(netflix_df["type"] == "Movie")]
netflix_subset
```

```python
# Tirando algumas colunas desnecessarias do dataframe
netflix_movies = netflix_subset.loc[:,["title","country","genre","release_year","duration"]]
netflix_movies
```

```python
# Filtrando filmes pequenos abaixo de 60 seg
short_movies = netflix_movies[(netflix_movies["duration"] < 60)]
short_movies[:20]
```

```python
# Adicionando Cores aos generos
colors = []
for lab, row in netflix_movies.iterrows():
    if (row['genre'] == 'Documentaries'):
        colors.append('Red')
    elif (row['genre'] == 'Children'):
        colors.append('Pink')
    elif (row['genre'] == 'Stand-Up'):
        colors.append('Blue')
    else:
        colors.append('Green')
```

```python
# Gerando grafico
fig = plt.figure(figsize=(12, 8))
plt.scatter(netflix_movies.release_year,netflix_movies.duration, c=colors)

# Adicionando os titulos
plt.title("Movie Duration by Year of Release")
plt.xlabel("Release year")
plt.ylabel("Duration (min)")
plt.show()
```

## O que podemos concluir?
Podemos concluir que os filmes no geral não estão diminuindo sua duração, apenas o volume de filmes aumentou e consequentemente vai ter a criação de mais videos curtos, como tambem mais filmes longos.
